Segmentasi Pelanggan (Clustering - KMeans)

Bisa mengetahui kelompok pelanggan berdasarkan pola belanja mereka (misalnya: High Spender, Average, Low Spender).

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Pilih fitur untuk clustering
X = df[['Order Item Quantity', 'Sales']].dropna()

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Gunakan metode KMeans dengan 3 klaster
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['customer_segment'] = kmeans.fit_predict(X_scaled)

# Visualisasi hasil clustering
plt.figure(figsize=(10,6))
sns.scatterplot(x=X['Order Item Quantity'], y=X['Sales'], hue=df['customer_segment'], palette='viridis')
plt.title('Segmentasi Pelanggan Berdasarkan Jumlah Pembelian & Sales')
plt.xlabel('Order Item Quantity')
plt.ylabel('Total Sales')
plt.show()
```

### **AI-Powered Predictions: Forecasting Penjualan**

Prediksi Penjualan Bulanan dengan ARIMA

Menunjukkan pemahaman bisnis yang lebih dalam
Bisa dijadikan rekomendasi strategi bisnis untuk mengantisipasi permintaan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Pastikan 'order date (DateOrders)' bertipe datetime
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')

# Agregasi data penjualan bulanan
df['order_month'] = df['order date (DateOrders)'].dt.to_period('M')  # Format PeriodIndex
df_sales_monthly = df.groupby('order_month')['Sales'].sum()

# Konversi index ke datetime dan tetapkan frekuensi agar kompatibel dengan ARIMA
df_sales_monthly.index = pd.to_datetime(df_sales_monthly.index.astype(str), format='%Y-%m')
df_sales_monthly = df_sales_monthly.asfreq('MS')  # Set frequency explicitly

# Uji stasioneritas dengan ADF Test
result = adfuller(df_sales_monthly.dropna())
print("ADF Statistic:", result[0])
print("p-value:", result[1])  # Jika p-value > 0.05, data tidak stasioner

# Model ARIMA dengan enforce_stationarity=False
model = ARIMA(df_sales_monthly, order=(5,1,0), enforce_stationarity=False)
model_fit = model.fit()

# Prediksi 6 bulan ke depan
forecast = model_fit.forecast(steps=6)

# Buat index untuk forecast agar cocok dengan df_sales_monthly
forecast_index = pd.date_range(start=df_sales_monthly.index[-1], periods=7, freq='MS')[1:]

# Plot hasil prediksi
plt.figure(figsize=(12,6))
plt.plot(df_sales_monthly, label='Historical Sales')
plt.plot(forecast_index, forecast, label='Forecasted Sales', linestyle='dashed', color='red')
plt.title('Prediksi Penjualan 6 Bulan ke Depan')
plt.xlabel('Waktu')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

## **Machine Learning**

### Prediksi Penjualan Menggunakan Regresi

Implementasi Decision Tree Regression

Perusahaan bisa memprediksi penjualan berdasarkan faktor-faktor yang tersedia dan melakukan strategi penjualan yang lebih tepat.



```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Pilih fitur dan target
features = ['Order Item Quantity', 'Order Item Discount Rate', 'Order Item Profit Ratio']
target = 'Sales'

# Hapus nilai kosong
df_ml = df[features + [target]].dropna()

# Split data
X_train, X_test, y_train, y_test = train_test_split(df_ml[features], df_ml[target], test_size=0.2, random_state=42)

# Inisialisasi dan latih model
model = DecisionTreeRegressor(max_depth=5)
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi model
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5  # Menghindari error pada sklearn < 0.22

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

## **Clustering Pelanggan dengan K-Means**

Implementasi K-Means untuk Segmentasi Pelanggan

Perusahaan bisa menargetkan pelanggan premium dengan strategi pemasaran yang lebih agresif.
Menyesuaikan promosi khusus untuk setiap segmen pelanggan.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Pilih fitur untuk clustering
X = df[['Order Item Quantity', 'Sales']].dropna()

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Gunakan metode KMeans dengan 3 klaster
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['customer_segment'] = kmeans.fit_predict(X_scaled)

# Visualisasi hasil clustering
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x=X['Order Item Quantity'], y=X['Sales'], hue=df['customer_segment'], palette='viridis')
plt.title('Segmentasi Pelanggan Berdasarkan Jumlah Pembelian & Sales')
plt.xlabel('Order Item Quantity')
plt.ylabel('Total Sales')
plt.show()
```

## **Prediksi Keterlambatan Pengiriman (Classification)**

Implementasi Random Forest Classifier

Perusahaan bisa mengidentifikasi pesanan yang berisiko terlambat sebelum pengiriman terjadi.
Bisa melakukan intervensi lebih awal untuk memperbaiki proses pengiriman.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Pilih fitur dan target
features = ['Order Item Quantity', 'Order Item Discount Rate', 'Sales']
target = 'Late_delivery_risk'

# Hapus nilai kosong
df_ml = df[features + [target]].dropna()

# Split data
X_train, X_test, y_train, y_test = train_test_split(df_ml[features], df_ml[target], test_size=0.2, random_state=42)

# Inisialisasi dan latih model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prediksi
y_pred = model.predict(X_test)

# Evaluasi model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Akurasi: {accuracy}')
print(report)
```
